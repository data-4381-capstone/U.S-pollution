```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data= pd.read_csv('/content/Cleaned_Feature-Engineered_Pollution_Data.csv')
```

```python
# Count the distribution of pollution severity
severity_counts = data['Pollution_Severity'].value_counts()

# Plot
severity_counts.plot(kind='bar', figsize=(8, 6))
plt.title("Distribution of Pollution Severity Levels")
plt.xlabel("Severity Levels")
plt.ylabel("Count")
plt.show()
```

```python
# Plot histogram with KDE for each pollutant
for pollutant in pollutants:
    plt.figure(figsize=(8, 6))
    sns.histplot(data[pollutant], kde=True, bins=30)
    plt.title(f"Distribution of {pollutant} Levels")
    plt.xlabel(f"{pollutant} Levels")
    plt.ylabel("Frequency")
    plt.show()
```

```python
# Compute correlation matrix
corr_matrix = data[pollutants].corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Pollutants")
plt.show()
```

```python
# Boxplot for each pollutant by pollution severity
for pollutant in pollutants:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Pollution_Severity', y=pollutant, data=data, palette="Set3")
    plt.title(f"{pollutant} Levels by Pollution Severity")
    plt.xlabel("Pollution Severity")
    plt.ylabel(f"{pollutant} Levels")
    plt.show()
```

```python
# Proportions of pollution severity
severity_proportions = severity_counts / severity_counts.sum()

# Plot
severity_proportions.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8), startangle=90)
plt.title("Proportions of Pollution Severity Levels")
plt.ylabel("")
plt.show()
```

```python
# Scatter plot for each pair of pollutants
for i in range(len(pollutants)):
    for j in range(i+1, len(pollutants)):
        plt.figure(figsize=(8, 6))
        sns.scatterplot(x=data[pollutants[i]], y=data[pollutants[j]])
        plt.title(f"Relationship Between {pollutants[i]} and {pollutants[j]}")
        plt.xlabel(pollutants[i])
        plt.ylabel(pollutants[j])
        plt.show()
```

```python
# Monthly average pollutant levels
monthly_avg = data.groupby('Month')[pollutants].mean()
monthly_avg.plot(kind='bar', figsize=(12, 6))
plt.title("Monthly Average Pollutant Levels")
plt.xlabel("Month")
plt.ylabel("Average Levels")
plt.legend(title="Pollutants")
plt.show()
```

```python
# Create pivot table for pollutants by month and pollutant
for pollutant in pollutants:
    seasonal_data = data.pivot_table(values=pollutant, index='Month', columns='Year', aggfunc='mean')

    plt.figure(figsize=(10, 6))
    sns.heatmap(seasonal_data, annot=True, fmt=".1f", cmap="YlGnBu")
    plt.title(f"Seasonal Heatmap of {pollutant} Levels")
    plt.xlabel("Year")
    plt.ylabel("Month")
    plt.show()
```
